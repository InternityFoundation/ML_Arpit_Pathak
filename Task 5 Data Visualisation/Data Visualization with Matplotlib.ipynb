```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Matplotlib

# Line Graph

```python
df = pd.read_csv("mobile_sales.csv")
df.head()
```

```python
plt.plot(df["Year"] , df["Quantity Sold"] , marker = "o")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("Mobile Sales")
plt.show()
```

# Bar Graph

```python
plt.bar(df["Year"] , df['Quantity Sold'] )
plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("Mobile Sales")
plt.xticks(df["Year"])
plt.show()
```

# Histogram

```python
da = pd.read_csv("500_Person_Gender_Height_Weight_Index.csv")
da.head()
```

```python
plt.hist(da["Weight"] , bins = 20 , histtype='bar', color = 'b')
plt.yticks(range(0,50,3))
plt.grid()
plt.show()
```

# Scatter plot

```python
plt.scatter(da["Height"] , da["Weight"] , c=da['Index'] , cmap = "viridis" )
plt.xlabel("Height(in cms)")
plt.ylabel("Weight(in kgs)")
plt.title("Body Mass Index")
plt.grid()
plt.show()
```
