```python
import pandas as pd
import numpy as np
```

```python
#importing the dataset covid-19
df = pd.read_csv('COVID-19.csv')
df.head()
```

```python
df.info()
```

```python
#categorizing the continuous variables age and temperature
bins= [0,10,20,55,100]
labels = ['Kids','Teen','Adult','Old Age']
Age = pd.cut(df['age'], bins=bins, labels=labels, right=False)
df["Age"]=Age
df=df.drop(['age'],axis=1)

bins= [96,98.6,102,110]
labels = ['normal','fever','high fever']
Temp = pd.cut(df['body temperature'], bins=bins, labels=labels, right=False)
df["Temp"]=Temp
df=df.drop(['body temperature'],axis=1)

df.head()
```

```python
print(df['Age'].value_counts())
print(df['Temp'].value_counts())
```

```python
#applying label encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Age'] = le.fit_transform(df['Age'])
df['Temp'] = le.fit_transform(df['Temp'])

print(df['Age'].value_counts())
print(df['Temp'].value_counts())
```

```python
df.head()
```

```python
#splitting the inportant features to x and y

X =df.drop(['Corona result' , 'Sno' , 'heart disease','kidney disease' , 'change in appetide' , 'lung disease' , 'weakness' ] , axis=1)
y = df['Corona result']
```

```python
#train and test split
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X , y , random_state = 1 , test_size = 0.20)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier  
model = DecisionTreeClassifier()  
model.fit(x_train, y_train)
DecisionTreeClassifier(class_weight=None,criterion='gini ',min_impurity_decrease=0.0,min_samples_leaf=1, min_samples_split=2,min_weight_fraction_leaf=0.0, presort=False, splitter='best')
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
#accuracy for training data
print(accuracy_score(y_train,model.predict(x_train)))
```

```python
print(confusion_matrix(y_train,model.predict(x_train)))
```

```python
#accuracy for testing data
print(accuracy_score(y_test,model.predict(x_test)))
```

```python
print(confusion_matrix(y_test,model.predict(x_test)))
```

# Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier

covid = RandomForestClassifier(min_samples_leaf=3 , max_features= 1 , random_state= 2)
covid.fit(x_train, y_train)
```

```python
#accuracy for training data
print(accuracy_score(y_train,covid.predict(x_train)))
```

```python
print(confusion_matrix(y_train,covid.predict(x_train)))
```

```python
#accuracy for testing data
print(accuracy_score(y_test,covid.predict(x_test)))
```
